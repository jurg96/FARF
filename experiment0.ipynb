```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import util

%load_ext autoreload
%autoreload 2
```

```python
filename = '../data/Set5/head.png'

data = util.load_image(filename)
xh = data.mean(axis=2)

s = 0.2
xl = util.resize(xh, s=s)
xl_upsampled = util.resize(xl, 1/s)

X = [xh, xl, xl_upsampled]
fig, ax = plt.subplots(1,3, figsize=(18,9))
for i, axx in enumerate(ax):
    x = X[i]
    axx.imshow(x, cmap='Greys_r')
    print(x.shape)
ax[0].set_xlabel('Original')
ax[1].set_xlabel('LR with s={}'.format(s))
ax[2].set_xlabel('LR with s={}, upsampled'.format(s))
```

```python
plt.figure(figsize=(8,8))
plt.imshow(xh - xl_upsampled, cmap='Greys_r')
plt.xlabel('Difference between HR image and upsampled LR image')
```

```python
Fs = util.get_features(xh - xl_upsampled)

fig, axes = plt.subplots(2,3, figsize=(18,12))
for i, ax in enumerate(axes.flat):
    ax.imshow(Fs[:,:,i], cmap='Greys_r')
```
